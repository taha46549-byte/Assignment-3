```python
## Q2: Student Marks Dictionary with Percentage

Students = { 
    "Taha" : [92,75,74,68,70],
    "Ali" : [85,48,96,32,41],
    "Ayan" : [58,65,95,25,44]
}

## Calculate percentage for each student

for name in Students:
    marks = Students[name]
    total = sum(marks)
    percentage = (total/500)*100
    print(name, "Percentage",percentage)
```

```python
## Dictionary of team's goals

Goals = {
    "team 1" : 2,
    "team 2" : 3,
    "team 3" : 1,
    "team 4" : 0
}

## Displaying all keys

print("Keys: ",Goals.keys())

## Displaying all values

print("Values: ",Goals.values())

## Displaying all key-value pairs

print("Items: ",Goals.items())

## Updating a students score

Goals.update({"team 3":2})
print("Updated score: ",Goals)

## Removing one student using pop

Goals.pop("team 4")
print("After removing team 4: ",Goals)
```
