```python
## Q7: A function that prints the table of 2.

def table(number):
    for t in range(1,11):
        print(number,"x",t,"=",number*t)
table(2)
```

```python
## Q8: A function that takes one number and checks whether it is even or odd.

def even_or_odd(number):
    if number / 2 == 0:
        print("Even")
    else:
        print("Odd")
even_or_odd(7)
```

```python
## Q9: A function that takes one number and prints its multiplication table.

def table(num):
    for t in range(1,11):
        print(num*t)
table(10)
```

```python
## Q10: A function is_even() that takes a number and prints is even or odd.

def is_even(num):
    if num % 2 == 0:
        print("Even")
    else:
        print("Odd")
is_even(8)
```
