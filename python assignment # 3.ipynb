```python
# Task 7: Mark sheet with grades
m1 = input("Maths: ")
m2 = input("Physics: ")
m3 = input("Computer: ")
m4 = input("Urdu: ")
m5 = input("English: ")

# Calculate total marks and percentage
obtain_marks = int(m1) + int(m2) + int(m3) + int(m4) + int(m5)
percentage = obtain_marks/500*100

# Decide grade based on percentage
if percentage >= 80: grade = "A+" 
elif percentage >= 70: grade = "A" 
elif percentage >= 60: grade = "B" 
elif percentage >= 50: grade = "C" 
elif percentage >= 40: grade = "D" 
else: grade = "F"

# Display mark sheet 
print("\n--- Mark Sheet ---") 
print("Total Marks:", obtain_marks) 
print("Percentage:", round(percentage, 2), "%") 
print("Grade:", grade)

```
